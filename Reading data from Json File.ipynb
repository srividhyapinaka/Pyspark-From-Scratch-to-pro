```
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

### Single-Line JSON 
A single-line JSON is a compact format where the entire JSON structure is in one line.

```
df = spark.read.format('json')\
    .option("multiline", False)\
        .option("inferschema", True)\
        .load('/FileStore/tables/drivers.json')
```

```
df.display()
```

### Multi-Line JSON 
A multi-line JSON consists of multiple JSON objects, each on its own line.

```
df_nested_json = spark.read.format('json')\
    .option("multiline", True)\
        .option("inferschema", True)\
        .load('/FileStore/tables/sample_nested-1.json')
```

```
df_nested_json.display()
```
